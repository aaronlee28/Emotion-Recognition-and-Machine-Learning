## Facial Detection

```python
face_df = pd.read_csv("facial_detection/emotions_features_final.csv")
face_df
```

```python
face_plot = face_df.groupby(by='emotion').count()
face_plot.rename(columns={"Unnamed: 0" : "Count"}, inplace=True)
face_plot
```

```python
face_bar = face_plot['Count']
face_bar.plot.bar(rot=0)
```

## Voice 

```python
voice_df = pd.read_csv("voice_detection/data/mood_features.csv")
voice_df.drop(['code'], axis =1, inplace = True)
voice_df
```

```python
sample = voice_df.head(50)
sample.head(1)
sample['features']= sample['features'].astype(str).str[1:-1]
```

```python
voice_scatter = sample.plot.scatter(x='emotions',y='features', c='DarkBlue')
y_axis = voice_scatter.axes.get_yaxis()
y_axis.set_visible(False)
```
